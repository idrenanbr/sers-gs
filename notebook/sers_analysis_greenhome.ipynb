```python
import pandas as pd
import matplotlib.pyplot as plt
from src.iot_simulation import apply_iot_logic

# Load data
file = '../data/home_office_energy_30d.csv'
df = pd.read_csv(file)
df.head()
```

```python
# Total consumption per day
cons_day = df.groupby('date')['total_consumption_kwh'].sum()
cons_day.plot(kind='bar', figsize=(14,4))
plt.title('Consumo total por dia (kWh)')
plt.show()
```

```python
# Solar vs Grid
solar = df['from_solar_kwh'].sum()
grid = df['from_grid_kwh'].sum()
plt.bar(['Solar','Rede'], [solar, grid])
plt.title('Comparação Solar vs Rede (kWh)')
plt.show()
```
